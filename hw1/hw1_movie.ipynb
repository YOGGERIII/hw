# (2)

```python
import pandas as pd
df = pd.read_csv('IMDB-Movie-Data.csv')

df['Actors'] = df['Actors'].str.split('|').apply(lambda x: [i.strip() for i in x]).tolist() #把所有演員存成list
df.explode('Actors').groupby('Actors').mean().nlargest(1, 'Revenue (Millions)', keep='all')[['Revenue (Millions)']] #把所有演員列出來後以演員分群,取平均收益最高的
```

```python
import pandas as pd
df = pd.read_csv('IMDB-Movie-Data.csv')

df['Actors'] = df['Actors'].str.split('|').apply(lambda x: [i.strip() for i in x]).tolist()
df['Genre'] = df['Genre'].str.split('|').apply(lambda x: [i.strip() for i in x]).tolist() #把所有類型存成list
df.explode('Actors').explode('Genre').groupby('Actors')['Genre'].nunique().nlargest(2, keep='all') #把所有演員和所有類型列出來後以演員分群,取演過類型前二多的
```

# (6)

```python
import pandas as pd
import numpy as np
df = pd.read_csv('IMDB-Movie-Data.csv')

df['Actors'] = df['Actors'].str.split('|').apply(lambda x: [i.strip() for i in x]).tolist()
df = df.explode('Actors').groupby('Actors')['Year'].agg([np.max, np.min]) #把所有演員列出來後以演員分群,取最大和最小年份
df['gap'] = df['amax'] - df['amin'] #創新欄gap為最大年份減最小年份
df.nlargest(3, 'gap', keep='all') #取gap前三大的演員
```
